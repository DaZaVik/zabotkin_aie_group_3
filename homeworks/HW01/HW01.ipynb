```python
import numpy as np

a = np.arange(1, 9)
z = np.zeros(3)
o = np.ones(7)
ar = np.arange(0, 10, 2)
r = np.linspace(0, 1, 5)

b = np.array([[1, 2, 3],
              [4, 5, 6]])

print("a:", a)
print("z:", z)
print("o:", o)
print("ar:", ar)
print("r:", r)
print("b:\n", b)

print("\n--- shape / dtype ---")
print("a.shape:", a.shape, "| a.dtype:", a.dtype)
print("b.shape:", b.shape, "| b.dtype:", b.dtype)
print("r.shape:", r.shape, "| r.dtype:", r.dtype)

x = np.arange(10)
print("\nМассив x:", x)
print("x[0] =", x[0])
print("x[-1] =", x[-1])
print("x[2:7] =", x[2:7])
print("x[::2] =", x[::2])

m = np.arange(16).reshape(4, 4)
print("\nМатрица m:\n", m)
print("Элемент [1, 2]:", m[1, 2])
print("Вторая строка:", m[1, :])
print("Третий столбец:", m[:, 2])
print("Подматрица 2x2:\n", m[1:3, 1:3])

print("\n--- Простые операции ---")
print("a      :", a)
print("a + 5  :", a + 5)
print("a * 2  :", a * 2)
print("a ** 2 :", a ** 2)

print("\nСложение массивов:")
print("ar:", ar)
print("ar + a[:5]:", ar + a[:5])

print("\n--- Статистики для a ---")
print("sum(a):", a.sum())
print("mean(a):", a.mean())
print("min(a):", a.min())
print("max(a):", a.max())
print("std(a):", a.std())

print("\n--- Статистики для m ---")
print("sum(m):", m.sum())
print("mean(m):", m.mean())
print("min(m):", m.min())
print("max(m):", m.max())
print("mean axis=0:", m.mean(axis=0))
print("mean axis=1:", m.mean(axis=1))
```
